```python
import numpy as np
import pandas as pd
import os
```

## Фичи товара

```python
os.chdir('C:/Users/rustem.kamilyanov/trainee/rec')
items = pd.read_csv('items.csv', header=0)
items['товар'] = [ i.replace("'", '').replace(",", '') for i in items['товар'] ]
items = items.fillna(0)
items = pd.get_dummies(items, columns=['тип', 'подтип'])

# стандартизируем
items['цена'] = [ (x-items['цена'].mean())/np.std(items['цена']) for x in items['цена'] ]

print(items.shape)
items.head()
```

## Фичи клиента

```python
users = pd.read_csv('features_clust.csv')
users = users.drop(['Unnamed: 0', 'smoking', 'name', 'Unnamed: 0.1'], axis=1)
users.head()
```

## user_item пустая

```python
items_dict = dict.fromkeys( [i for i in items['товар']], np.zeros(1000) )
user_item = pd.DataFrame(items_dict)
user_item.insert(0, 'age', users['age'])
user_item.insert(1, 'clust', users['clust'])
user_item.insert(2, 'childs', users['childs'])
user_item.index.name = 'user_id'
user_item.head()
```

## user_item заполняем покупками в зависимости от возраста юзера и наличия детей

```python
items_list = items['товар']
age_list_child = items.columns[2:4]
age_list = items.columns[4:7]

np.random.seed(14)
for i in items_list:
    for a in age_list:
        if (items[ items['товар']==i ][a] == 1).bool():
            # определяем возрастную группу для товара
            age1, age2 = a.split('-')
            
            # берем всех юзеров соответствующих текущей возрастной группе товара
            idx = user_item[i][ (user_item['age'] > int(age1)) & (user_item['age'] <= int(age2)) ].index
            
            # рандомного генерим количество покупок товара от 0 до 3 для получившейся группы юзеров
            # (большее количество покупок генерится с меньшей вероятностью)
            size = len(idx)
            user_item.loc[idx, i] = np.random.choice(range(0, 4), size, p=[0.8, 0.15, 0.04, 0.01], replace=True)
        else:
            pass
    
    for c in age_list_child:
        if (items[ items['товар']==i ][c] == 1).bool():
            
            # берем всех юзеров с наличием детей
            idx = user_item[i][user_item['childs'] != 'Нет'].index
            
            # рандомного генерим количество покупок детских товаров от 0 до 3 для получившейся группы юзеров
            # (большее количество покупок генерится с меньшей вероятностью)
            size = len(idx)
            user_item.loc[idx, i] = np.random.choice(range(0, 4), size, p=[0.8, 0.15, 0.04, 0.01], replace=True)
        else:
            pass

user_item.to_csv('user_items.csv')
user_item.head()
            
            
            
```
