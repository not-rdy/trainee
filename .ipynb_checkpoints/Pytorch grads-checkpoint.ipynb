```python
import os
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('dark_background')

import torch
```

```python
from sklearn.datasets import load_boston
from IPython.display import clear_output
# data
x = load_boston()['data'][:, -1]
y = load_boston()['target']

# data to tensor
x = torch.tensor(x / x.max(), dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)

# weights
w = torch.rand(1, dtype=torch.float32, requires_grad=True)
b = torch.rand(1, dtype=torch.float32, requires_grad=True)

# learning rate
lr = 0.05
for i in range(1000):
    # предикт
    y_hat = w * x + b
    
    # считаем лосс и вектор градиента
    loss = torch.mean((y - y_hat)**2)
    loss.backward()
    
    # корректируем веса с lr (градиентный шаг)
    w.data -= lr * w.grad.data
    b.data -= lr * b.grad.data
    
    # обнуляем элементны вектора градиента по каждому весу
    w.grad.data.zero_()
    b.grad.data.zero_()
    
    if i % 5 == 0:
        clear_output(True)
        plt.scatter(x.numpy(), y.numpy())
        plt.plot(x.numpy(), y_hat.data.numpy(), color='orange', linewidth=5)
        plt.show()
        print(f'loss: {loss.item()}')
    
    
```
